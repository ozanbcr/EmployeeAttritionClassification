```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import mean_absolute_error,accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
y=df["Attrition"]
```

```python
x=df.drop("Attrition",axis=1)
```

```python
y
```

```python
#Training model
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0,shuffle=True,stratify=y)
```

```python
scaler=StandardScaler()
scaler.fit(x_train)
x_train=pd.DataFrame(scaler.transform(x_train),index=x_train.index,columns=x_train.columns)
x_test=pd.DataFrame(scaler.transform(x_test),index=x_test.index,columns=x_test.columns)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC,LinearSVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Support Vector Machine (Linear Kernel)": LinearSVC(),
    "Support Vector Machine (RBF Kernel)": SVC(),
    "Neural Network": MLPClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier()
}

for name, model in models.items():
    model.fit(x_train, y_train)
    print(name + " trained.")
```

```python
for name, model in models.items():
    print(name + ": {:.2f}%".format(model.score(x_test,y_test) * 100))
```

```python
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
for name, model in models.items():
  predictions = model.predict(x_test)
  report = classification_report(y_test, model.predict(x_test))
  accuracy = accuracy_score(y_test, predictions)
  print('\nModel:',name)
  print("The accuracy is " ,accuracy)
  print("\n", report)
  plot_confusion_matrix(model, x_test, y_test)
  plt.title('Confusion Matrix for '+ name)  
  plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
#  Logistic Regression parameter optimization
params_LR = {
    'penalty' : ['l1','l2'], 
    'C'       : np.logspace(-3,3,7),
    'solver'  : ['newton-cg', 'lbfgs', 'liblinear'],
}
```

```python
logreg = LogisticRegression()
clf = GridSearchCV(logreg,                    # model
                   param_grid = params_LR,   # hyperparameters
                   scoring='accuracy',        # metric for scoring
                   cv=10)                     # number of folds
```

```python
clf.fit(x_train,y_train)
```

```python
print("Tuned Hyperparameters :", clf.best_params_)
print("Accuracy :", clf.best_score_)
```

```python
logreg = LogisticRegression(C = 0.1, 
                            penalty = 'l2', 
                            solver = 'newton-cg')
logreg.fit(x_train,y_train)
y_pred = logreg.predict(x_test)
print("Accuracy:",logreg.score(x_test, y_test))
```

```python
report = classification_report(y_test, logreg.predict(x_test))
accuracy = accuracy_score(y_test, logreg.predict(x_test))
print('\nModel Logistic Regression:',)
print("The accuracy is " ,accuracy)
print("\n", report)
plot_confusion_matrix(logreg, x_test, y_test)
plt.title('Confusion Matrix for Logistic Regression')  
plt.show()
```

```python
#Decision Tree parameter opt.

params_DT={'criterion' : ['gini','entropy'],
           'max_depth' : range(1,10),
           'min_samples_split' : range(1,10),
           'min_samples_leaf'  : range(1,5)}
```

```python
dt=DecisionTreeClassifier()
grid_DT=GridSearchCV(dt,
                     param_grid=params_DT,
                     scoring='accuracy',
                     cv=10)
grid_DT.fit(x_train,y_train)
```

```python
print("Tuned Hyperparameters :", grid_DT.best_params_)
```

```python
grid_DT.best_score_
```

```python
dtree=DecisionTreeClassifier(criterion='entropy',max_depth=2,min_samples_leaf=1,min_samples_split=2)
```
